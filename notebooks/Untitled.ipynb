```python
from nagini import Star, Spot, Planet, STSP
```

```python
import numpy as np

# Define a planet: 
planet = Planet(per=2, inc=90, a=4, rp=0.1, 
                ecc=0, w=90, lam=0, t0=0)

# Define a star: 
star = Star(planet=planet, rotation_period=27, 
            inc_stellar=90, spot_contrast=0.7, 
            u=[0.6, 0.2], rho_s=1)
```

```python
import astropy.units as u

# Define times to compute light curve: 
times = np.linspace(-0.1, 0.1, 1000)

# Define a spot: 
spots = [Spot(0*u.deg, 0*u.deg, 0.2)]
```

```python
with STSP(times, star, spots) as s: 
    lc = s.generate_lightcurve(normalize_oot=True)

lc.plot()
```
